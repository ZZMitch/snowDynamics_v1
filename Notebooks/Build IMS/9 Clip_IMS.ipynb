# Clip all IMS products, removing 0s outside
- This currently does all years, may need to adjust for future single year additions

```python
import glob
import os
from tqdm import tqdm
import xarray as xr
import rioxarray as rxr
import geopandas as gpd
```

```python
#roi = r'C:\Users\mbonney\OneDrive - NRCan RNCan\Projects\HBL\Extent\HBL_CanadaROI.shp'
#roi = r"C:\Users\mbonney\OneDrive - NRCan RNCan\Projects\UtilityData\Study Areas\Canada\Boundary\CanVec\CanVec_Dissolve3979_Buffer10km_Erase_Edit_FillHoles.shp"
roi = gpd.read_file(roi, engine = 'pyogrio')
roi
```

```python
input = r'C:\Users\mbonney\OneDrive - NRCan RNCan\Projects\High Resolution Snow Dynamics\IMS (1 km)\Outputs\interannual'
input_ID = 'pSnowFree_1824'
#variable = 'snow_status'
output_ID = 'Canada'
output = r'C:\Users\mbonney\OneDrive - NRCan RNCan\Projects\High Resolution Snow Dynamics\IMS (1 km)\Outputs\interannual'
```

```python
# Find all relevant IMS files
files = [file for file in glob.glob(os.path.join(input, './**/*.tif'), recursive = True) if input_ID in file]
files
```

```python
for file in tqdm(files):
    da = rxr.open_rasterio(file).astype('float32')
    da = da.rio.clip(roi.geometry.values)
    da.rio.write_nodata(-999, encoded = True, inplace = True)
    #da.rio.to_raster(os.path.join(output, 'IMS_' + variable + '_' + file[-18:-4] + '_' +  output_ID + '.tif'))
    da.rio.to_raster(os.path.join(output, file.split('\\')[-1][:-4] + '_' + output_ID + '.tif'))
```

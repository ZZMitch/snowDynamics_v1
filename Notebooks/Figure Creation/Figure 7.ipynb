# Plot

```python
fig = plt.figure(figsize = (6.5, 6.5 * 5/6))

### A. End dates
# 2018-2019
axA_1819 = plt.subplot2grid((5, 6), (0, 0))

end_1819.plot(ax = axA_1819, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axA_1819.set_title('$\\bf{A.}$End', fontsize = 10)
axA_1819.set_ylabel('2018-2019', fontsize = 8)
axA_1819.set_box_aspect(1)

# 2019-2020
axA_1920 = plt.subplot2grid((5, 6), (1, 0))

end_1920.plot(ax = axA_1920, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axA_1920.set_ylabel('2019-2020', fontsize = 8)
axA_1920.set_box_aspect(1)

# 2020-2021
axA_2021 = plt.subplot2grid((5, 6), (2, 0))

end_2021.plot(ax = axA_2021, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)
# Color bar in open space later

axA_2021.set_ylabel('2020-2021', fontsize = 8)
axA_2021.set_box_aspect(1)

# 2021-2022
axA_2122 = plt.subplot2grid((5, 6), (3, 0))

end_2122.plot(ax = axA_2122, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axA_2122.set_ylabel('2021-2022', fontsize = 8)
axA_2122.set_box_aspect(1)

# 2022-2023
axA_2223 = plt.subplot2grid((5, 6), (4, 0))

end_2223.plot(ax = axA_2223, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axA_2223.set_ylabel('2022-2023', fontsize = 8)
axA_2223.set_box_aspect(1)

### B. End uncertainty
# 2018-2019
axB_1819 = plt.subplot2grid((5, 6), (0, 1))

end_u_1819.plot(ax = axB_1819, add_labels = False, xticks = [], yticks = [], vmin = 0, vmax = 15, add_colorbar = False)
axB_1819.set_title('$\\bf{B.}$Unc.', fontsize = 10)

axB_1819.set_box_aspect(1)

# 2019-2020
axB_1920 = plt.subplot2grid((5, 6), (1, 1))

end_u_1920.plot(ax = axB_1920, add_labels = False, xticks = [], yticks = [], vmin = 0, vmax = 15, add_colorbar = False)

axB_1920.set_box_aspect(1)

# 2020-2021
axB_2021 = plt.subplot2grid((5, 6), (2, 1))

end_u_2021.plot(ax = axB_2021, add_labels = False, xticks = [], yticks = [], vmin = 0, vmax = 15, add_colorbar = False)

axB_2021.set_box_aspect(1)

# 2021-2022
axB_2122 = plt.subplot2grid((5, 6), (3, 1))

end_u_2122.plot(ax = axB_2122, add_labels = False, xticks = [], yticks = [], vmin = 0, vmax = 15, add_colorbar = False)

axB_2122.set_box_aspect(1)

# 2022-2023
axB_2223 = plt.subplot2grid((5, 6), (4, 1))

end_u_2223.plot(ax = axB_2223, add_labels = False, xticks = [], yticks = [], vmin = 0, vmax = 15, add_colorbar = False)

axB_2223.set_box_aspect(1)

### C. Weighted mean
# Snow End Date
axC_end = plt.subplot2grid((5, 6), (0, 2), colspan = 2, rowspan = 2)

end_mn.plot(ax = axC_end, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axC_end.set_title('$\\bf{C.}$Weighted Mean', fontsize = 10)
axC_end.set_box_aspect(1)

# Quality
axC_q = plt.subplot2grid((5, 6), (3, 2), colspan = 2, rowspan = 2)

end_q_mn.plot(ax = axC_q, add_labels = False, xticks = [], yticks = [], vmin = 0.75, vmax = 1, add_colorbar = False)

axC_q.set_box_aspect(1)

### D. Best value
# Snow End Date
axD_end = plt.subplot2grid((5, 6), (0, 4), colspan = 2, rowspan = 2)

end_bv.plot(ax = axD_end, add_labels = False, xticks = [], yticks = [], vmin = 115, vmax = 155, cmap = 'coolwarm_r', add_colorbar = False)

axD_end.set_title('$\\bf{D.}$Best Value', fontsize = 10)
axD_end.set_box_aspect(1)

# Year
axD_y = plt.subplot2grid((5, 6), (2, 5))

colors = ['#E41A1C', '#377EB8', '#4DAF4A', '#984EA3', '#FF7F00']
scale = [2018.5, 2019.5, 2020.5, 2021.5, 2022.5, 2023.5]
cmap = mpl.colors.ListedColormap(colors)
norm = mpl.colors.BoundaryNorm(scale, len(colors))  

end_bvy.plot(ax = axD_y, add_labels = False, xticks = [], yticks = [], cmap = cmap, norm = norm, add_colorbar = False)

axD_y.set_box_aspect(1)

# Quality
axD_q = plt.subplot2grid((5, 6), (3, 4), colspan = 2, rowspan = 2)

end_bvq.plot(ax = axD_q, add_labels = False, xticks = [], yticks = [], vmin = 0.75, vmax = 1, add_colorbar = False)

axD_q.set_box_aspect(1)

### Colorbars
ax_cbars = plt.subplot2grid((5, 6), (2, 2), colspan = 3)
ax_cbars.set_axis_off()

# Uncertainty
cmap = mpl.cm.viridis
norm = mpl.colors.Normalize(vmin = 0, vmax = 15)
cax = ax_cbars.inset_axes([0.02, 0.1, 0.05, 0.85])
cbar = fig.colorbar(mpl.cm.ScalarMappable(norm = norm, cmap = cmap), cax = cax, orientation = 'vertical', extend = 'max')
cbar.ax.tick_params(labelsize = 8)
cbar.set_label(label = 'Uncertainty\n(Days)', size = 8)

# End
cmap = mpl.cm.coolwarm_r
norm = mpl.colors.Normalize(vmin = 115, vmax = 155)
cax = ax_cbars.inset_axes([0.3, 0.85, 0.4, 0.1])
cbar = fig.colorbar(mpl.cm.ScalarMappable(norm = norm, cmap = cmap), cax = cax, orientation = 'horizontal', extend = 'both')
cbar.ax.tick_params(labelsize = 8, length = 2) # pad, 
cbar.set_label(label = 'End Date (Since Dec 31)', size = 8, labelpad = 2)

# Quality
cmap = mpl.cm.viridis # Could change to plasma, inferno, magma, cividis if confusing with unc...
norm = mpl.colors.Normalize(vmin = 0.75, vmax = 1)
cax = ax_cbars.inset_axes([0.3, 0.05, 0.4, 0.1])
cbar = fig.colorbar(mpl.cm.ScalarMappable(norm = norm, cmap = cmap), cax = cax, orientation = 'horizontal', extend = 'min')
cbar.ax.tick_params(labelsize = 8, length = 2)
cbar.set_label(label = 'Quality', size = 8, labelpad = 3)
cbar.ax.xaxis.set_ticks_position('top')
cbar.ax.xaxis.set_label_position('top')

# Best Value Year
cmap = mpl.colors.ListedColormap(colors)
norm = mpl.colors.BoundaryNorm(scale, len(colors))  
cax = ax_cbars.inset_axes([0.77, 0.12, 0.05, 0.75])
cbar = fig.colorbar(mpl.cm.ScalarMappable(norm = norm, cmap = cmap), cax = cax, orientation = 'vertical', ticks = [2019, 2020, 2021, 2022, 2023])
cbar.ax.tick_params(labelsize = 8)
cbar.set_label(label = 'Year', size = 8)

plt.tight_layout(h_pad = 0, w_pad = -0.1) #(h_pad = 0.5, w_pad = 0.5)

fig.get_figure().savefig('Figure 8_1.tif', dpi = 600, bbox_inches = 'tight', pil_kwargs = {'compression': 'tiff_lzw'})
```

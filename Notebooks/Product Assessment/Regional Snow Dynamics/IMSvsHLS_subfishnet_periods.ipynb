## Graphical Plot
- Stacked percentage bar, % agreement - weighted on count (HLS periods on X)

```python
weights = (hls_stats['count'] / hls_stats['count'].max()) * (ims_stats['count'] / ims_stats['count'].max())
weights.name = 'weight'
weights
```

```python
df = pd.DataFrame(index = hls_stats.index)
df['HLS_mode'] = hls_stats['mode']
df['HLS_mean'] = hls_stats['mean'] # Helpful to see disagreements
df['IMS_mode'] = ims_stats['mode']
df['IMS_mean'] = ims_stats['mean'] # Helpful to see disagreements
df = df.merge(weights, left_index = True, right_index = True)
df['offset'] = df['HLS_mean'] - df['IMS_mean']
df['offset_abs'] = abs(df['offset'])
df['agreement'] = ''
df.loc[(df['IMS_mode'] == 1) & (df['HLS_mode'] == 1), 'agreement'] = 'TN' # True value (IMS) is 1 and HLS is 1
df.loc[(df['IMS_mode'] == 1) & (df['HLS_mode'] > 1), 'agreement'] = 'FP' # True value (IMS) is 1 and HLS is 2
df.loc[(df['IMS_mode'] > 1) & (df['HLS_mode'] > 1), 'agreement'] = 'TP' # True value (IMS) is 2 and HLS is 2
df.loc[(df['IMS_mode'] > 1) & (df['HLS_mode'] == 1), 'agreement'] = 'FN' # True value (IMS) is 2 and HLS is 1
df
```

```python
HLS_periods = ('One', 
               'Multiple')

hls1_all = len(df[df['HLS_mode'] == 1])
hls1_ims1 = len(df[(df['HLS_mode'] == 1) & (df['IMS_mode'] == 1)])
hls1_ims2 = len(df[(df['HLS_mode'] == 1) & (df['IMS_mode'] > 1)])

hls2_all = len(df[df['HLS_mode'] > 1])
hls2_ims2 = len(df[(df['HLS_mode'] > 1) & (df['IMS_mode'] > 1)])
hls2_ims1 = len(df[(df['HLS_mode'] > 1) & (df['IMS_mode'] == 1)])

status = {'Agree': np.array([hls1_ims1 / hls1_all, hls2_ims2 / hls2_all]),
          'Disagree': np.array([hls1_ims2 / hls1_all, hls2_ims1 / hls2_all])}

width = 0.9

fig, ax = plt.subplots()
bottom = np.zeros(2)

for boolean, status in status.items():
    p = ax.bar(HLS_periods, status, width, label=boolean, bottom=bottom)
    bottom += status

#ax.set_title("Number of penguins with above average body mass")
#ax.legend(loc="upper right")

plt.show()

# Blue = Agree
# Orange = Disagree
```

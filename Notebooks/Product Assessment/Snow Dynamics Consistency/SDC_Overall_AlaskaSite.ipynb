```python
yr1 = hls_2122
yr2 = hls_2223
```

```python
%%time
print('From xarray: ' + str(xr.corr(yr1, yr2, dim = ['x', 'y']).load().item()))
```

```python
%%time
# If null in either arr, set to null in both
yr1 = xr.where((yr2.isnull() | yr1.isnull()), np.nan, yr1)
yr2 = xr.where((yr2.isnull() | yr1.isnull()), np.nan, yr2)

# Flatten
yr1 = yr1.to_numpy().flatten()
yr2 = yr2.to_numpy().flatten()

# Drop null
yr1 = yr1[~np.isnan(yr1)]
yr2 = yr2[~np.isnan(yr2)]

print('From scipy:' + str(pearsonr(yr1, yr2)[0]) + ', pval: ' + str(pearsonr(yr1, yr2)[1]))
```
